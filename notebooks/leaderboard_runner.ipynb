# Home Credit Leaderboard Runner

Thin notebook wrapper for the existing pipeline code in `src/hc_risk/`.

Adjust the run flags below, then execute the notebook top to bottom.

Start with the sampled cached run first. A full `FORCE=True` rebuild on CPU can take hours.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / "src" / "hc_risk").exists() and (ROOT.parent / "src" / "hc_risk").exists():
    ROOT = ROOT.parent
if not (ROOT / "src" / "hc_risk").exists():
    raise RuntimeError("Run this notebook from the repo root or from notebooks/.")

SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

ROOT
```

```python
from pprint import pprint

from hc_risk.config import PipelineConfig
from hc_risk.leaderboard import run_leaderboard
```

```python
FORCE = False
TUNING_TRIALS = 1
LEADERBOARD_SAMPLE_ROWS = 50000

config = PipelineConfig(
    leaderboard_tuning_trials=TUNING_TRIALS,
    leaderboard_sample_rows=LEADERBOARD_SAMPLE_ROWS,
)
config
```

```python
print(f"Starting leaderboard run with force={FORCE}, sample_rows={LEADERBOARD_SAMPLE_ROWS}, tuning_trials={TUNING_TRIALS}")
metrics = run_leaderboard(config, force=FORCE)
pprint(metrics)
```

```python
print("Final blend ROC AUC:", metrics["blend"]["roc_auc"])
print("Base rank blend ROC AUC:", metrics["base_blend"]["roc_auc"])
print("Artifacts written under:", config.outputs_dir)
```
